```r
#Lectura de Datos
ubicacion<-'https://raw.githubusercontent.com/Emax1/R/master/water_potability.csv'
datos<-read.csv(ubicacion,header = TRUE,sep=',',stringsAsFactors = TRUE)
```

```r
datos
```

```r
dim(datos)
names(datos)
str(datos)
```

```r
summary(datos)
```

```r
#eliminar NA
datos2<-na.omit(datos)
dim(datos2)
summary(datos)
```

```r
tri<-datos2$Trihalomethanes
hist(tri)
```

```r
tri_media<-round(mean(tri),2)
#media
tri_media<-round(median(tri),2)
tri_p2<-quantile(tri,0.20)
tri_q1<-quantile(tri,0.25)
tri_q2<-quantile(tri,0.5)
tri_q3<-quantile(tri,0.75)
tri_d1<-quantile(tri,0.10)
tri_d2<-quantile(tri,0.20)
tri_d4<-quantile(tri,0.40)
tri_d5<-quantile(tri,0.50)
tri_d9<-quantile(tri,0.90)
```

```r
tri_r<-max(tri)-min(tri)
tri_iqr<-tri_q3-tri_q1
tri_var<-var(tri)
tri_sd<-sd(tri)
tri_sd/tri_media*100
```

```r
boxplot(tri,horizontal = TRUE)
```

```r
tri<-datos2$Trihalomethanes
fi<-table(tri)
n<-length(tri)
hi<-fi/n
pi1<-hi*100
cbind(fi,hi,pi1)
pie(pi1)
barplot(pi1)
```
